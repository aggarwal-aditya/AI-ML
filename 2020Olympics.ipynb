```python
import pandas as pd
df=pd.read_csv('Athletes.csv')
```

```python
df.head()
```

## Which Country had most number of Athletes

```python
grouped=df.groupby('NOC').count()['Name']
```

```python
grouped.sort_values(ascending=False).head()
```

```python
grouped.sort_values(ascending=False).head().plot.bar(figsize=(20,5))
```

## Which Athlete took part in maximum sports

```python
df.groupby('Name').count()['Discipline']
```

```python
df.groupby('Name').count()['Discipline'].sort_values(ascending=False).head(10)
```

## Which Sport had maximum athletes

```python
df.groupby('Discipline').count()['Name']
```

```python
df.groupby('Discipline').count()['Name'].sort_values(ascending=False).head().plot.bar()
```

## China had max athletes in which discipline?


```python
china=df[df['NOC']=="People's Republic of China"];
```

```python
china.groupby('Discipline').count()['Name'].sort_values(ascending=False).plot.bar()
```

## Difference between Number of Athletes from China and Italy

```python
print(len(df[df['NOC']=="People's Republic of China"])-len(df[df['NOC']=="Italy"]))
```

## Nmber of Male & Female Athletes in Shooting

```python
df[df['Discipline']=='Shooting']
```

```python
pd=df.groupby("Discipline").sum()['Female']
print(len(pd))
```

```python
y=0
z=0;
for x in range(len(pd)):
  y=y+pd[x]
  z=z+df.groupby('Discipline').sum()['Male'][x]
print("Female Athletes:",y)
print("Male Athletes:",z)
```
